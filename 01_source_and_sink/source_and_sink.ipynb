# Potential flow theory: source and sink

Potential flow is described by Laplace's equation:

$$\nabla^2\phi=0$$

$$u=\frac{\partial \phi}{\partial x}, \quad v=\frac{\partial \phi}{\partial y}, \quad w=\frac{\partial \phi}{\partial z}$$

The velocity components in radial coordinates $\left(r,\theta\right)$, where the angle $\theta$ is $\tan^{-1}(y/x)$, are given by:

$$u_r\left(r,\theta\right) = \frac{\sigma}{2\pi r} \text{,} \qquad u_\theta\left(r,\theta\right)=0$$

where $\sigma$ represents the source *strength*. According to the *stream function* $\psi$, we can know that:

$$u = \frac{\partial\psi}{\partial y} = \frac{\sigma}{2\pi}\frac{x-x_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$

and

$$v = -\frac{\partial\psi}{\partial x} = \frac{\sigma}{2\pi}\frac{y-y_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$

### Importing dependencies

```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

### Generating a mesh grid

```python
N = 50                              # Number of points
x_start, x_end = -2.0, 2.0          # Boundaries in the x-direction
y_start, y_end = -1.0, 1.0          # Boundaries in the y-direction
x = np.linspace(x_start, x_end, N)  # Array with x-coordinates
y = np.linspace(y_start, y_end, N)  # Array with y-coordinates

X, Y = np.meshgrid(x, y)            # Mesh grid
```

```python
# Plotting the mesh
x_size = 10
y_size = (y_end-y_start)/(x_end-x_start)*x_size
plt.figure(figsize=(x_size, y_size))
plt.title("Mesh grid")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.scatter(X, Y, s=8, color="#CD2305", marker='o', linewidth=0)
plt.show()
```

### Source flow

```python
strength_source = 5.0               # Strength of the source
x_source, y_source = -1.0, 0.0      # Location of the source

# Computing the velocity field on the mesh grid
u_source = strength_source/(2*np.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = strength_source/(2*np.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
```

```python
# Plotting the streamlines from the source
x_size = 10
y_size = (y_end-y_start)/(x_end-x_start)*x_size
plt.figure(figsize=(x_size, y_size))
plt.title("Source flow")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_source, v_source, density=2, color="#4B3FE8", linewidth=1, arrowsize=1, arrowstyle="-|>")
plt.scatter(x_source, y_source, s=80, color="#CD2305", marker='o', linewidth=0)
plt.show()
```

### Sink flow

```python
strength_sink = -5.0            # Strength of the sink
x_sink, y_sink = 1.0, 0.0       # Location of the sink

# Computing the velocity field on the mesh grid
u_sink = strength_sink/(2*np.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = strength_sink/(2*np.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
```

```python
# Plotting the streamlines to the sink
x_size = 10
y_size = (y_end-y_start)/(x_end-x_start)*x_size
plt.figure(figsize=(x_size, y_size))
plt.title("Sink flow")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_sink, v_sink, density=2, color="#4B3FE8", linewidth=1, arrowsize=1, arrowstyle="-|>")
plt.scatter(x_sink, y_sink, s=80, color="#CD2305", marker='o', linewidth=0)
plt.show()
```

### Source-sink pair

```python
# Computing the velocity field on the mesh grid
u_pair = u_source + u_sink
v_pair = v_source + v_sink

# Plotting the streamlines
x_size = 10
y_size = (y_end-y_start)/(x_end-x_start)*x_size
plt.figure(figsize=(x_size, y_size))
plt.title("Source-sink flow")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_pair, v_pair, density=2, color="#4B3FE8", linewidth=1, arrowsize=1, arrowstyle="-|>")
plt.scatter([x_source, x_sink], [y_source, y_sink], s=80, color="#CD2305", marker='o', linewidth=0)
plt.show()
```
